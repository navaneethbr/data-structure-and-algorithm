## operator overloading

```python
class Book:
    def __init__(self, pages):
        self.pages = pages
        
    def __add__(self, others):
        return self.pages+others.pages
    
    # for more passing more than 2 argument for operator overloading
    def __sub__(self, others):
        s = self.pages-others.pages
        b = Book(s)
        return b
    
    def __str__(self):
        return str(self.pages)
    
b1 = Book(200)
b2 = Book(300)
b3 = Book(2000)
print(b3-b1-b2)
# print(b3)
```

## operator overloading between 2 class

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        
    def __mul__(self, others):
        return self.salary*others.days*12

class Timesheet:
    def __init__(self, name, days):
        self.name = name
        self.days = days
        
e = Employee("Navaneeth", 2000)
t = Timesheet("Navaneeth", 30)

print(e*t)
```

## method overriding 

```python
class Car:
    def propert(self):
            print("This is a car")
    def brand(self):
        print("Honda")
class ModCar(Car):
    def brand(self):
        super().brand()
        print("Hyundai")
o = ModCar()
o.brand()
```

## MRO(Method Resolution Order)
### C3 Algorithm 
#### Example


##### mro(P) = P + Merge(mro(X), mro(Y), mro(C), XYC)
##### mro(P) = P + Merge(XABO, YBCO, CO, XYC)
##### mro(P) = P + X + Merge(ABO, YBCO, CO, YC)
##### mro(P) = P + X + A + Merge(BO, YBCO, CO, YC)
##### mro(P) = P + X + A + Y + Merge(BO, BCO, CO, C)
##### mro(P) = P + X + A + Y + B + Merge(O, CO, CO, C)
##### mro(P) = P + X + A + Y + B + C + Merge(O, O, O)
##### mro(P) = P + X + A + Y + B + C + O 

```python
class Person:
    def details(self, name, age):
        self.name = name
        self.age = age
    def display(self):
        print(self.name)
        print(self.age)
class Student(Person):
    def __init__(self, name, age, rollno, marks):
        super().details(name, age)
        self.rollno = rollno
        self.marks = marks
class Teacher(Student):
    def __init__(self, name, age, salary, subject):
        super().details(name, age)
        self.salary = salary 
        self.subject = subject
s = Student('navaneeth', 22, 3, 100)
t = Teacher('Reddy', 42, 50000, 'python')
t.display()
s.display()
```
